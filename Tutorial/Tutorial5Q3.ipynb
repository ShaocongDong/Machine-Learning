#### Question 3: Bias and Variance

```python
# import library
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
import math
```

```python
# set parameters for experiments
def f(x): 
    return np.sin(np.pi*x)
y_value = lambda x: np.array((f(x[0]), f(x[1])))
```

```python
def generate_training_data(num):
    '''
    This function generates a list of tuples to be the training data
    :param num: number of experiments
    :return: the list of tuples of training set
    '''
    result = []
    for i in range(num):
        x = rd.rand(2)*2-1
        y = y_value(x)
        result += [(x, y),]
    return result
```

```python
def fit_single_tuple(dataTrain):
    '''
    Look at my notes for details of derivation 
    Only the result is shown here
    :param dataTrain: training date of one tuple
    :return: the parameter a
    '''
    x1, x2, y1, y2 = dataTrain[0][0], dataTrain[0][1], dataTrain[1][0], dataTrain[1][1]
    return ((x1*y1 + x2*y2) / (x1**2 + x2**2))
```

```python
def experiment(num):
    # set up the plots
    h = 0.01
    x = np.arange(-1, 1, h)
    plt.plot(x, f(x), "b-")
    dataTrain = generate_training_data(num)
    a_fit_result = []
    for i in range(num):
        a_fit = fit_single_tuple(dataTrain[i])
        a_fit_result.append(a_fit)
        plt.plot(x, a_fit*x, "m-", alpha=0.1)
    plt.show()
    print("Variance of fitted a: " + str(np.var(a_fit_result)))
    return a_fit_result
    
```

```python
a_fit_result = experiment(100)
```

```python
def fit_single_tupple_h2(dataTrain):
    x1, x2, y1, y2 = dataTrain[0][0], dataTrain[0][1], dataTrain[1][0], dataTrain[1][1]
    a = ((y2-y1) / (x2-x1))
    b = ((x2*y1 - x1*y2) / (x2-x1))
    return a, b


    
```
