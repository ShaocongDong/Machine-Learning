```python
import sklearn as skt
import sklearn.datasets as data
import sklearn.linear_model as lm
import numpy as np
import matplotlib.pyplot as pl
```

```python
lr = lm.base.LinearRegression()
data_received = data.load_boston(return_X_y = True)
crim = data_received[0][:,0]
target = data_received[1]
```

```python
crim = crim.reshape(506,1)
target = target.reshape(506,1)
regr = lr.fit(crim, target)
predict = regr.predict(crim)
```

```python
pl.scatter(crim, target, color = 'red', edgecolors=(0, 0, 0))
pl.plot(crim, predict, color='blue', linewidth=4)
pl.xticks()
pl.yticks()
pl.show()
```

```python
all_feature_data = data_received[0]
target = data_received[1]
lr_t = lm.base.LinearRegression()
```

```python
regr_t = lr_t.fit(all_feature_data, target)
print(regr_t.coef_) # examine coefficients of regression
```

```python
# plot the first two for reference
predict_t = regr_t.predict(all_feature_data)
predict_t = predict_t.reshape(506,1)
raw_crim = all_feature_data[:,0].reshape(506,1)
raw_age = all_feature_data[:,1].reshape(506,1)
```

```python
# clear original plot - 1
pl.clf()
pl.cla()
```

```python
pl.scatter(raw_crim, target, color = 'red', edgecolors=(0, 0, 0))
pl.plot(raw_crim, predict_t, color='blue', linewidth=1)
pl.xticks()
pl.yticks()
pl.show()
```

```python
# clear original plot - 2
pl.clf()
pl.cla()
```

```python
pl.scatter(raw_age, target, color = 'red', edgecolors=(0, 0, 0))
pl.plot(raw_age, predict_t, color='blue', linewidth=1)
pl.xticks()
pl.yticks()
pl.show()
```

```python
# clear original plot - 3
pl.clf()
pl.cla()
```

```python
combine_age = np.column_stack((raw_age, predict_t))
combine_age = combine_age[combine_age[:,0].argsort()]

combine_target = np.column_stack((raw_age, target))
combine_target = combine_target[combine_age[:,0].argsort()]

pl.scatter(combine_target[:,0], combine_target[:,1], color = 'red', edgecolors=(0, 0, 0))
pl.plot(combine_age[:,0], combine_age[:,1], color='blue', linewidth=1)
pl.xticks()
pl.yticks()
pl.show()
```
