* Question3: Visualizing the data and using KNN

```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as pl
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets as ds
from sklearn.neighbors import KNeighborsClassifier as knc
from sklearn.linear_model import LogisticRegression as lg
```

```python
dataset = ds.load_iris(return_X_y=True)
dependent = dataset[1]
predictors = dataset[0]
x1, x2, x3, x4 = predictors[:,0], predictors[:,1], predictors[:,2], predictors[:,3]
#print(ds.load_iris()['DESCR'])
```

```python
def scatter_plot_features (feature_1, feature_2, color_chosen, label_chosen):
    pl.scatter(feature_1, feature_2, c=color_chosen, label=label_chosen)
    pl.show()

def scatter_plot_features_dependent (feature_1, feature_2, dependent, color_map_set):
    pl.scatter(feature_1, feature_2, c=dependent, cmap=color_map_set, edgecolor = 'k')
    pl.show()

def sample_error_ratio(predict, dependent):
    N = predict.shape[0]
    sum_error = 0
    for i in range(N):
        if predict[i]!=dependent[i]:
            sum_error+=1
    return sum_error/N

def try_Neighbor_with_error(maxK, predictors, dependent):
    for i in range(1, maxK+1):
        neigh = knc(n_neighbors=i)
        neigh.fit(predictors, dependent)
        predict = neigh.predict(predictors)
        print("Sample error ratio with number of neighbors equal to: " + str(i) 
              +" is: " + str(sample_error_ratio(predict, dependent)))
    return neigh
    
def try_Logistic_with_error(predictors, dependent):
    logis = lg()
    logis.fit(predictors, dependent)
    predict = logis.predict(predictors)
    print("Sample error ratio with Logistic Ratio equal to: " +
         str(sample_error_ratio(predict, dependent)))
    return logis
```

```python
scatter_plot_features_dependent(x1, x2, dependent, pl.cm.Set2)
scatter_plot_features_dependent(x3, x4, dependent, pl.cm.Set3)
```

```python
try_Neighbor_with_error(20, predictors=predictors, dependent=dependent)
```
