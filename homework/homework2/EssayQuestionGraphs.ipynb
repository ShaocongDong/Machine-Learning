```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x1 = np.vstack((1,2,2,0,1,0)).reshape(1,6)[0]
x2 = np.vstack((1,2,0,1,0,0)).reshape(1,6)[0]
y = np.vstack((1,1,1,-1,-1,-1)).reshape(1,6)[0]
```

```python
pl.scatter(x1,x2,c=y)
x = np.arange(0, 2, 0.1)
yprime = lambda x: 1.5-x
plt.plot(x, yprime(x))
plt.show()
```
