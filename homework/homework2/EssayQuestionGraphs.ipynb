```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x1 = np.vstack((1,2,2,0,1,0)).reshape(1,6)[0]
x2 = np.vstack((1,2,0,1,0,0)).reshape(1,6)[0]
y = np.vstack((1,1,1,-1,-1,-1)).reshape(1,6)[0]
```

```python
pl.scatter(x1,x2,c=y)
pl.show()
```

```python
pl.scatter(x1,x2,c=y)
x_lin = np.arange(0, 2, 0.1)
boundary_function = lambda x1: 1.5-x1
boundary_function_1 = lambda x1: 1-x1
boundary_function_2 = lambda x1: 2-x1
y_lin = boundary_function(x_lin)
y_lin_1 = boundary_function_1(x_lin)
y_lin_2 = boundary_function_2(x_lin)
plt.plot(x_lin, y_lin)
plt.plot(x_lin, y_lin_1, 'r--')
plt.plot(x_lin, y_lin_2, 'r--')
plt.show()
```
